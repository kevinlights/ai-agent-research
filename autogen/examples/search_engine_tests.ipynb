```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://cn.bing.com/search?q=opensearch&first=1")
soup = BeautifulSoup(response.content, "html.parser")

links = []
try:
    li_list = soup.select("#b_results")[0].select("li")
    for li in li_list:
        try:
            a = li.select(".b_tpcn")[0].select("a")[0]
            if 'href' in a.attrs:
                link = a['href']
                links.append(link)
        except:
            print(f"parse link error: {li}")
except:
    print("server error, please wait for some minutes and retry")
len(links), links
```

```python
from bs4 import BeautifulSoup
def parse_bing_results(page_content: str):
    results = []
    try:
        soup = BeautifulSoup(page_content, "html.parser")
        li_list = soup.select("#b_results")[0].select("li")
        for li in li_list:
            try:
                a = li.select(".b_tpcn")[0].select("a")[0]
                link = li.select("h2")[0].select("a")[0].text
                snippet = li.select(".b_caption")[0].select("p")[0].text
                if "href" in a.attrs:
                    link = a["href"]
                    results.append({
                        "title": "",
                        "link": link,
                        "snippet": snippet,
                        "body": "",
                    })
            except:
                # print(f"parse item error: {li}")
                pass
    except Exception as e:
        # print(e)
        pass
    return results
parse_bing_results(page_content)
```
