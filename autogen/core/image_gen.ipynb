```python
from models import llama3_sd_prompt, chat
from zhipu import draw_image_with_b64
from autogen_core import (
    Image,
)

model_client = llama3_sd_prompt()

prompt = (
    "1girl, cute, vibrant young girl, curly brown hair, standing alone in a lush forest"
)

res = await chat(prompt, model_client)
print(res.content)


b64str = draw_image_with_b64(res.content)
image = Image.from_base64(b64str)
image = Image.from_pil(image.image.resize((256, 256)))
display(image.image)  # type: ignore
```
